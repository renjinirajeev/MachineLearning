```python
#Seaborn - data visualization
import seaborn as sns
df=sns.load_dataset('tips')
df
```

```python
#correlation

df.corr()
```

```python
#correlation can be reperesent graphically  using heatmap()
sns.heatmap(df.corr())
```

```python
# join plot used for numerical columns

sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')  # kind='reg'
```

```python
sns.jointplot(x='tip',y='size',data=df)
```
