**IMPORTING DATA AND LIBRARIES**

```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/stroke.csv")
df
```

**PREPROCESSING**

```python
# To print first five columns

df.head()
```

```python
# To print last five columns

df.tail()
```

```python
# dropping the irrelevant columns

df.drop(['id'],axis=1,inplace=True)
df
```

**CHECKING FOR MISSING VALUES**

```python
df.isna().sum()
```

```python
# Filling the missing values


df['bmi']=df['bmi'].fillna(df['bmi'].mean())
df.isna().sum()
```

**IDENTIFYING THE DATATYPES AND ENCODING**

```python
df.dtypes
```

```python
# Encoding


from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['gender']=lb.fit_transform(df['gender'])
df['ever_married']=lb.fit_transform(df['ever_married'])
df['work_type']=lb.fit_transform(df['work_type'])
df['Residence_type']=lb.fit_transform(df['Residence_type'])
df['smoking_status']=lb.fit_transform(df['smoking_status'])
df.dtypes
```

**SEPERATING X AND Y VARIABLES**

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

**SPLIT THE DATA INTO TRAINING AND TESTING DATA**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)


# Input training data


x_train
```

```python
# Input testing data

x_test
```

```python
# Output Training data

y_train
```

```python
# Output testing data

y_test
```

**NORMALIZATION USING MINMAXSCALER**

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

**MODEL CREATION USING KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

**PERFORMANCE EVALUATION**

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

**CALCULATING ACCURACY**

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
label=[0,1]
cmd=ConfusionMatrixDisplay(cm,display_labels=label)
cmd.plot()
```

**CLASSIFICATION REPORT**

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
