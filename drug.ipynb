```python
df1=pd.get_dummies(df[['Sex','BP','Cholesterol','Drug']],drop_first=True)
df1
```

```python
df2=pd.concat([df,df1],axis=1)
df2
```

```python
df2.drop(['Sex','BP','Cholesterol','Drug'],axis=1,inplace=True)
df2
```

```python
df.dtypes
```

```python
df2.dtypes
```

```python
x=df2
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  data=KMeans(n_clusters=i,init='k-means++',random_state=42)
  data.fit(x)
  wcss.append(data.inertia_)
```

```python
print(wcss)
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('No of clusters')
plt.ylabel('Variance')
plt.title('Elbow Point')
```

```python
data1=KMeans(n_clusters=4,init='k-means++',random_state=42)
data1.fit(x)
y=data1.predict(x)
y
```

```python
# performance evaluation   : silhouette_score

from sklearn.metrics import silhouette_score
print(silhouette_score(x,y))

```

```python
x=df2.iloc[:,:-1].values
x
```

```python
y=df2.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
score=accuracy_score(y_test,y_pred)
score
```
