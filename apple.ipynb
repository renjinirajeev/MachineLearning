```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/apple_quality.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isna().sum()
```

```python
df.drop(['A_id'],axis=1,inplace=True)
df.isna().sum()
```

```python
df['Size']=df['Size'].fillna(df['Size'].mean())
df['Weight']=df['Weight'].fillna(df['Weight'].mean())
df['Sweetness']=df['Sweetness'].fillna(df['Sweetness'].mean())
df['Crunchiness']=df['Crunchiness'].fillna(df['Crunchiness'].mean())
df['Juiciness']=df['Juiciness'].fillna(df['Juiciness'].mean())
df['Ripeness']=df['Ripeness'].fillna(df['Ripeness'].mean())
df['Quality']=df['Quality'].fillna(df['Quality'].mode()[0])
df.isna().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Acidity']=lb.fit_transform(df['Acidity'])
df.dtypes
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
 x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
labels=['bad','good']
cmd=ConfusionMatrixDisplay(cm,display_labels=labels)
cmd.plot()
```

```python
from sklearn.naive_bayes import BernoulliNB
base=BernoulliNB()
base.fit(x_train,y_train)
y_pred=base.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
labels=['bad','good']
cmd=ConfusionMatrixDisplay(cm,display_labels=labels)
cmd.plot()
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
score=accuracy_score(y_test,y_pred)
score
```
