```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/Laptops.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isna().sum()
```

```python
m=df['Storage'].mode()[0]
df['Storage'].fillna(m,inplace=True)
df.isna().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Brand']=lb.fit_transform(df['Brand'])
df['Model Name']=lb.fit_transform(df['Model Name'])
df['Processor']=lb.fit_transform(df['Processor'])
df['Operating System']=lb.fit_transform(df['Operating System'])
df['Storage']=lb.fit_transform(df['Storage'])
df['RAM']=lb.fit_transform(df['RAM'])
df['Screen Size']=lb.fit_transform(df['Screen Size'])
df['Touch_Screen']=lb.fit_transform(df['Touch_Screen'])
df.dtypes
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=2)
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
score=accuracy_score(y_test,y_pred)
score
```
