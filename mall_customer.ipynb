```python
df.drop(['CustomerID','Genre'],axis=1,inplace=True)
df
```

```python
x=df
x
```

```python
from sklearn.cluster import KMeans
wcss=[]            # empty list        var1 var2 var3...............var10
for i in range(1,11):       #i=1  i=2.............10
  data=KMeans(n_clusters=i,init='k-means++',random_state=42)    #cluster=1        #Kmeans object creat chaithu  init = initialization of kmeans algorithm
  data.fit(x)      # x fit chaith nokkune
  wcss.append(data.inertia_)      # inertia_  : function to find the variance   variance list append
```

```python
print(wcss)
```

```python
plt.plot(range(1,11),wcss)        # x= range(1,11)  from 1 to 10
plt.xlabel('No of clusters')
plt.ylabel('Variance')
plt.title('Elbow Points')
```

```python
from sklearn.cluster import KMeans
data1=KMeans(n_clusters=6,init='k-means++',random_state=42)
data1.fit(x)
y=data1.predict(x)
y
```

```python
x['cluster']=y
x
```

```python
x1=df.iloc[:,:-1].values
x1
```

```python
y1=df.iloc[:,-1].values
y1
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
score=accuracy_score(y_test,y_pred)
score
```
