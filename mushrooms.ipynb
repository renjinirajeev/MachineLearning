```python
df1=pd.get_dummies(df[['cap-shape','cap-surface','cap-color','bruises','odor','gill-attachment','gill-spacing','gill-size','gill-color','stalk-shape','stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring','stalk-color-below-ring','veil-type','veil-color','ring-number','ring-type','spore-print-color','population','habitat']],drop_first=True)
df1
```

**SPLIT X AND Y**

```python
x=df1.values
x
```

```python
y=df['class'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

**SCALING**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
param={'n_neighbors':[3,5,7,9],'weights':['uniform','distance']}
clf=GridSearchCV(knn,param,cv=10,scoring='accuracy')
clf.fit(x_train,y_train)
# param={'n_neighbors':[3,5,7,9],'weights':['uniform','distance']}
# clf=GridSearchCV(knn1,param,cv=10,scoring='accuracy')
# clf.fit(x_train,y_train)
```

```python
print(clf.best_params_)
```

```python
knn=KNeighborsClassifier(n_neighbors=3,weights='uniform')
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay, classification_report
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
print('score is:',accuracy_score(y_test,y_pred))
```

```python
labels=['e','p']
cmd=ConfusionMatrixDisplay(cm,display_labels=labels)
cmd.plot()
```

```python
print(classification_report(y_test,y_pred))
```

**Naive Bayes**

```python
from sklearn.naive_bayes import BernoulliNB
base=BernoulliNB()
```

```python
param={'alpha':np.logspace(-3,3,num=7),'binarize': np.linspace(0.1, 1.0, num=10)}
clf=GridSearchCV(base,param,cv=10,scoring='accuracy')
clf.fit(x_train,y_train)
```

```python
print(clf.best_params_)
```

```python
base=BernoulliNB(alpha=0.001,binarize=1)
base.fit(x_train,y_train)
y_pred=base.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay, classification_report
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
print('score is:',accuracy_score(y_test,y_pred))
```

```python
labels=['e','p']
cmd=ConfusionMatrixDisplay(cm,display_labels=labels)
cmd.plot()
```

```python
print(classification_report(y_test,y_pred))
```
