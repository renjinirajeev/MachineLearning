```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/company_sales_data (2).csv')
df
```

```python
#1


x=df['month_number']
y=df['total_profit']
plt.plot(x,y)
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Month Vs Profit')
plt.xticks(x)
plt.yticks([100000,200000,300000,400000,500000])
```

```python
#2


# x=df['month_number']
# y=df['total_profit']
plt.plot(x,y,linestyle='dotted',color='r',marker='o',mfc='r',linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.xticks(x)
plt.yticks([100000,200000,300000,400000,500000])
```

```python
#3


face_cream=df['facecream']
face_wash=df['facewash']
toothpaste=df['toothpaste']
bathingsoap=df['bathingsoap']
shampoo=df['shampoo']
moisturizer=df['moisturizer']

# plt.plot(x,face_cream,marker='s',mfc='g',color='r')
# plt.plot(x,face_wash,marker='s',mfc='g',color='y')
# plt.plot(x,toothpaste,marker='s',mfc='g',color='k')
# plt.plot(x,bathingsoap,marker='s',mfc='g',color='m')
# plt.plot(x,shampoo,marker='s',mfc='g',color='b')
# plt.plot(x,moisturizer,marker='s',mfc='g',color='c')
# plt.xlabel('Month')
# plt.ylabel('Product')
# plt.title('Month Vs Product')
# plt.xticks(x)
# plt.yticks([1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,11000,12000])


lst=[face_cream,face_wash,toothpaste,bathingsoap,shampoo,moisturizer]
for i in lst:
  plt.plot(x,i,marker='s',mfc='g',color='r')
plt.xlabel('Month')
plt.ylabel('Product')
plt.title('Month Vs Product')
plt.xticks(x)
plt.yticks([1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,11000,12000])


```

```python
#4


# month=df['month_number']
toothpaste=df['toothpaste']
plt.scatter(x,toothpaste)
plt.xlabel('Month')
plt.ylabel('Toothpaste')
plt.title('Month Vs Toothpaste')
```

```python
#5


plt.bar([a-0.25 for a in x],face_cream,width=0.25,label='face_cream',align='edge')
plt.bar([a+0.25 for a in x],face_wash,width=0.25,label='face_wash',align='edge')
plt.xlabel('Month')
plt.xticks(x)
plt.ylabel('sales data')
plt.legend(loc='upper right')
plt.grid(linewidth=1,linestyle='dashed')
plt.title('Total Production of Facewash & Facecream')
```

```python
#6

profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(y)
plt.xlabel('profit')
plt.ylabel('month')
plt.title('histrogram of profit in each month')
plt.xticks(profit_range)
plt.yticks(x)
```

```python
# 7


label=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
k=[face_cream.sum(),face_wash.sum(),toothpaste.sum(),bathingsoap.sum(),shampoo.sum(),moisturizer.sum()]
plt.pie(k,labels=label,autopct='%1.1f%%')
plt.legend(loc='lower right')
```
